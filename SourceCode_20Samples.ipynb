```python
#import Gaussian Multivariate PDF from scipy.stats
from scipy.stats import multivariate_normal

#use Gaussian Model (Probability function) for Class1
C1_M= multivariate_normal(mean=class1_mean, cov=covMatrix1)
#print(C1_M.pdf(Class1.T))

#use Gaussian Model (Probability function) for Class2
C2_M= multivariate_normal(mean=class2_mean, cov=covMatrix2)
#print(C2_M.pdf(Class2.T))

#use Gaussian Model (Probability function) for Class3
C3_M= multivariate_normal(mean=class3_mean, cov=covMatrix3)
#print(C3_M.pdf(Class3.T))

#use Gaussian Model (Probability function) for Class4
C4_M= multivariate_normal(mean=class4_mean, cov=covMatrix4)
#print(C4_M.pdf(Class4.T))
```

```python
#Initialize predicted data as numpy array
predic_data = [[0 for c in range(356)] for r in range(211)]
predic_data = np.array(predic_data)

for i in range (0,211): 
    local_vect = []
    for j in range (0,356):
        #Save 6 dimensional vector of pixel value for each iteration in a local vector
        local_vect.append(r[i][j])
        local_vect.append(g[i][j])
        local_vect.append(b[i][j])
        local_vect.append(le[i][j])
        local_vect.append(fe[i][j])
        local_vect.append(nir[i][j])
        
        #Feed each pixel value (6D vector) to all 4 models and get Probability results
        c1_p = C1_M.pdf(local_vect)
        c2_p = C2_M.pdf(local_vect)
        c3_p = C3_M.pdf(local_vect)
        c4_p = C4_M.pdf(local_vect)
        
        #Determine the highest probability from the results
        high = max(c1_p,c2_p,c3_p,c4_p)
        
        #Fill predicted data array
        #Assign class to pixel value based on highest result
        if high == c1_p:
            predic_data[i][j] = 1
        elif high == c2_p:
            predic_data[i][j] = 2
        elif high == c3_p:
            predic_data[i][j] = 3
        elif high == c4_p:
            predic_data[i][j] = 4
            
        local_vect.clear()
        
```

```python
#print the prediction results
print(predic_data)
```

```python
#import confusion matrix from sklearn
from sklearn.metrics import confusion_matrix, classification_report

y_true = data
y_pred = predic_data

labels = ["1", "2", "3", "4"]

conf_mat_dict={}

#Create a dictionary to store original and predicted data for each class
for label_col in range(len(labels)):
    y_true_label = y_true[:, label_col]
    y_pred_label = y_pred[:, label_col]
    conf_mat_dict[labels[label_col]] = confusion_matrix(y_pred=y_pred_label, y_true=y_true_label)

#Print confusion matrix for each class
for label, matrix in conf_mat_dict.items():
    print("Confusion matrix for class {}:".format(label))
    print(matrix)
```

```python
#Generate classification report
clf_report_dict={}
for label_col in range(len(labels)):
    y_true_label_cr = y_true[:, label_col]
    y_pred_label_cr = y_pred[:, label_col]
    clf_report_dict[labels[label_col]] = classification_report(y_true=y_true_label_cr, y_pred=y_pred_label_cr)

#Print classification report
for label, clf_rpt in clf_report_dict.items():
    print("Classification Report for Class {}:".format(label))
    print(clf_rpt)
```
